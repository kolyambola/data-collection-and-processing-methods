1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests
from pprint import pprint
import json
```

```python
url = 'https://api.github.com/users/kolyambola/repos'
```

```python
response = requests.get(f'{url}')
```

```python
response.status_code
```

```python
response = response.json() 
```

```python
pprint(response)
```

```python
repositories = []

for repo in response:
    repositories.append(repo['name'])
```

```python
print(repositories)
```

```python
with open('user_list_repositories.json', 'w') as f:
    json_repo = json.dump(repositories, f)
```

2. Изучить список открытых API (https://www.programmableweb.com/category/all/apis). Найти среди них любое, требующее авторизацию (любого типа). Выполнить запросы к нему, пройдя авторизацию. Ответ сервера записать в файл.

```python
def get_data(url, lat, lon, appid):
    while True:
        url = f'{url}?lat={lat}&lon={lon}&appid={appid}'
        response = requests.get(url)
        if response.status_code == 200:
            print(url)
            break  
    return response.json()         
```

```python
lon = '-118.244'
```

```python
response = get_data(url_2, lat, lon, appid)
```
