```python
import requests
from lxml import html
from pprint import pprint
import pandas as pd
```

```python
url = 'https://dzen.ru/news/'
```

```python
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36 Edg/108.0.1462.54,gzip(gfe)'
```

```python
headers = {
    'User_Agent': user_agent 
}

params = {
    'sso_failed': ''
}
```

```python
def get_dom(url, headers=headers, params=params):
    response = requests.get(url, headers=headers, params=params)
    dom = html.fromstring(response.text)

    print(response.url)
    
    return dom
```

```python
def parser_dzen_news(dom):
    news = dom.xpath("//div[contains(@class,'mg-card ')]")
    dzen_news = []
    for new in news:
        news_source = new.xpath(".//a[@class='mg-card__source-link']/text()")[0]
        news_text = new.xpath(".//h2/a/text()")[0].replace('\xa0', ' ')
        news_link = new.xpath(".//h2/a/@href")[0]
        news_date = new.xpath(".//span[@class='mg-card-source__time']/text()")[0]
        news_dict = {
            'source': news_source,
            'text': news_text,
            'link': news_link,
            'date': news_date,
        }
        dzen_news.append(news_dict) 

    return dzen_news
```

```python
dom = get_dom(url, headers=headers, params=params)
```

```python
dzen_news = parser_dzen_news(dom)
dzen_news
```

```python
df = pd.DataFrame(dzen_news)
df.head()
```
